```python
import pandas as pd
```

```python
victims = pd.read_csv('data/victims.csv') 
victims_geo = pd.read_csv('data/victims_geo.csv') 
```

```python
victims_geo.head()
```

```python
victims_geo.shape
```

```python
victims_geo = victims_geo[victims_geo["victim"].str.contains("Adam Smith")==False]
```

```python
victims_geo = victims_geo[victims_geo["victim"].str.contains("Abraham Arellano")==False]
```

```python
victims_geo = victims_geo[victims_geo["victim"].str.contains("Robert Edwards")==False]
```

```python
victims_geo = victims_geo[((victims_geo['victim'].str.contains("Michael Brown")) & (victims_geo['victim_age'] !='18'))==False]
```

```python
victims_geo[victims_geo['victim']=='Michael Brown']
```

```python
victims_geo.shape
```

```python
victims_geo = victims_geo.groupby('victim').filter(lambda x : len(x)<2)
```

```python
victims_geo_new = victims.merge(victims_geo, left_on='victim', right_on='victim')
```

```python
victims_geo_new[victims_geo_new['victim']=='Michael Brown']
```

```python
victims_geo_new.to_csv('data/victims_geo_new.csv',index=False)
```
