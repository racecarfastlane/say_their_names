```python
import requests
import pandas as pd
```

```python
url = "https://mappingpoliceviolence.org/s/MPVDatasetDownload.xlsx"
r = requests.get(url, allow_redirects=True)
open('data/MPVDatasetDownload.xlsx','wb').write(r.content) 
```

```python
mpv_db_pk = pd.read_excel('data/MPVDatasetDownload.xlsx',sheet_name='2013-2020 Police Killings')
mpv_db_pkpd = pd.read_excel('data/MPVDatasetDownload.xlsx',sheet_name='2013-2020 Killings by PD')
mpv_db_pkstate = pd.read_excel('data/MPVDatasetDownload.xlsx',sheet_name='2013-2020 Killings by State')
```

```python
mpv_db_pk.head(10)
```

```python
mpv_db_pk = mpv_db_pk.rename(columns={"Victim\'s name": "victim_name", "Victim\'s age": "victim_age","Victim\'s gender": "victim_gender","Victim\'s race": "victim_race"})
```

```python
mpv_db_pk.head()
```

```python
droplist = [i for i in mpv_db_pk.columns if i.startswith('Unnamed')]
mpv_db_pk.drop(droplist,axis=1,inplace=True)
```

```python
mpv_db_pk.shape
```

```python
mpv_db_pk.to_csv('data/mpv_db_pk.csv',index=False)
mpv_db_pkpd.to_csv('data/mpv_db_pkpd.csv',index=False)
mpv_db_pkstate.to_csv('data/mpv_db_pkstate.csv',index=False)
```
