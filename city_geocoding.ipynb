```python
import pandas as pd
```

```python
cities = pd.read_csv('data/uscities.csv')  
mpv_db_pk = pd.read_csv('data/mpv_db_pk.csv')
```

```python
droplist = ['city_ascii','county_fips','county_name','population','density','source','military','incorporated','timezone',
           'ranking','zips','id']
cities.drop(droplist,axis=1,inplace=True)
```

```python
zips = pd.read_csv('data/us-zip-code-latitude-and-longitude.csv',sep=";")
```

```python
zips['Zip'] = zips['Zip'].fillna(0).astype(int)
```

```python
droplist = ['Timezone','Daylight savings time flag','geopoint']
zips.drop(droplist,axis=1,inplace=True)
```

```python
mpv_db_pk['Zipcode'] = mpv_db_pk['Zipcode'].fillna(0).astype(int)
```

```python
#First Merge
test1 = mpv_db_pk.merge(zips,how="left", left_on=['Zipcode'], right_on=['Zip'])
```

```python
test1.shape
```

```python
#First Merge dropping NAs
test_latlong = test1.dropna(subset=['Latitude'])
```

```python
test_latlong.shape
```

```python
#Slice from Test 1 where Latitude is null
test2 = test1[test1['Latitude'].isna()]   
```

```python
test2.shape
```

```python
#Second Merge-- Remaining NA's merged with cities dataframe
test_merge = test2.merge(cities,how="left", left_on=['City_x','State_x'], right_on=['city','state_id'])
```

```python
test_merge.shape
```

```python
test_merge[test_merge['lat'].isna()]
```

```python
test_merge = test_merge.dropna(subset=['lat'])
```

```python
test_merge.head(2)

#test_merge = test_merge.drop(columns=['Latitude', 'Longitude','city','state_id','state_name'])
```

```python
droplist = ['city', 'state_id', 'state_name', 'Latitude', 'Longitude']
test_merge.drop(droplist,axis=1,inplace=True)
```

```python
test_merge = test_merge.rename(columns={"lat": "Latitude", "lng": "Longitude"})
```

```python
test_merge.columns
```

```python
test_latlong.columns
```

```python
frames = [test_merge, test_latlong]

result = pd.concat(frames)
```

```python
result[result['Latitude'].isna()]
```

```python
result['index'] = range(1, len(result) + 1)
```

```python
result = result.set_index('index')
```

```python
result = result.rename(columns={"victim_name": "victim"})
```
